```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://en.wikipedia.org/wiki/List_of_Indian_states_and_union_territories_by_GDP_per_capita"
data=requests.get(url)
soup =BeautifulSoup(data.text,'html')
print(soup)
```

```python
table = soup.find('table',class_='wikitable sortable')
print(table)
```

```python
titles = table.find_all('th')
print(titles)
```

```python
table_head = [title.text.strip() for title in titles]
print(table_head)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns= table_head)
df
```

```python
column_data = table.find_all('tr')
print(column_data)
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]
    print(individual_row_data)
    
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data=[data.text.strip() for data in row_data]
    
print(individual_row_data)
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data =[data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```
